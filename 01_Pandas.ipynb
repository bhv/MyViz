```python
# convert 'fare' into float
df_titanic['fare'] = df_titanic['fare'].astype('float')
df_titanic['fare'].dtype
```

```python
# convert 'body' into numeric
df_titanic['body'] = pd.to_numeric(df_titanic['body'])
df_titanic['body'].dtype
```

```python
df_titanic.describe()
```

```python
# categorical columns
col_cat = df_titanic.select_dtypes(include=['object','category']).columns
pd.Series(col_cat)
```

```python
def describe_cat(df):
    col_cat = list(df_titanic.select_dtypes(include=['object','category']).columns)
    
    df.fillna(np.nan)
    for col in col_cat:
        print(f'\n-- Column name: {col}')
        #replace '?' by np.nan
        df[col] = [np.nan if x=='?' else x for x in df[col]]
        print(f'Total unique values: {df[col].nunique()}')
        print(f'Value counts:')
        print(df[col].value_counts())
        print(f'percent missing: {round(df[col].isna().sum() * 100/df[col].shape[0])}')
        
describe_cat(df_titanic)
        
    
```
